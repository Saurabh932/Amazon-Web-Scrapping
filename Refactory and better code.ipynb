```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import numpy as np
```

```python
# Function to extract Product Title
def get_title(soup):
    
    try:
        # Outer Tag Object
        title = soup.find("span", attrs = {'id':'productTitle'})
        
        # Inner NavipigatableString Obects
        title_value = title.text
        
        # Title as a String Value
        title_string = title_value.strip()
        
    except AttributeError:
        title_string = ""
        
    return title_string


# Function to extract product Price
def get_price(soup):
    
    try:
        # Outer Tag Object
        price = soup.find("span", attrs={"class":'a-price aok-align-center'}).find("span", attrs={"class":'a-offscreen'}).text
        
    except :
        price = ""
        
    return price


# Function to extract product Rating
def get_rating(soup):
    
    try:
        # Outer Tag Object
        rating = soup.find("i", attrs={'class':'a-icon a-icon-star a-star-4-5'}).text
        
    except AttributeError:
        rating = ""
        
    return rating



# Function to extract Avaliable Status
def get_avaliablity(soup):
    
    try:
        # Outer Tag Object
        avaliablity = soup.find("div", attrs={'id':'availability'})
        avaliablity = avaliablity.find("span").string.strip()
        
    except AttributeError:
        avaliablity = "NOt Avaliablity"
        
    return avaliablity
```

```python
if __name__ == '__main__':

    # Header for request
    HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36', 'Accept-Language': 'en-US, en;q=0.5'})
    
    # The Website URL
    URL = "https://www.amazon.in/s?k=playstation+5&sprefix=plays%2Caps%2C518&ref=nb_sb_ss_softlines-tsdoa-joint-contextual-iss_1_5"
    
    # HTTP Request'
    webpage = requests.get(URL, headers = HEADERS)
    
    # Soup Objects containing all data
    soup = BeautifulSoup(webpage.content, 'html.parser')
    
    # Fetch links as List of Tag Objects
    links = soup.find_all('a', attrs = {'class': 'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})
    
    # Store the links in the list
    links_list = []
    
    # Loop for extracting list from Tag Object
    for link in links:
        links_list.append(link.get('href'))
        
    d = {"title":[], "price":[], "rating":[], "availability":[]}
    
    
    # Loop for extracting product details from each links
    for link in links_list:
        new_webpage = requests.get("https://amazon.in" + link, headers=HEADERS)
        
        new_soup = BeautifulSoup(new_webpage.content, 'html.parser')
        
        
        # Function calls to display all necessary deatails of the product information
        d['title'].append(get_title(new_soup))
        d['price'].append(get_price(new_soup))
        d['rating'].append(get_rating(new_soup))
#         d['reviews'].append(get_reviews_count(new_soup))
        d['availability'].append(get_avaliablity(new_soup))
        
    amazon_df = pd.DataFrame.from_dict(d)
    amazon_df['title'].replace('', np.nan, inplace = True)
    amazon_df = amazon_df.dropna(subset=['title'])
    amazon_df.to_csv("amazon_data.csv", header=True, index = False)
```
